```python
# ── Imports & seed ──────────────────────────────────────────────────────────
import random
import numpy as np
import pandas as pd
from pathlib import Path

SEED = 42
random.seed(SEED)
np.random.seed(SEED)

# Relative paths only — never hardcode /Users/...
RAW_DIR    = Path("../../data/raw")
OUTPUT_DIR = Path("outputs")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
```

## 1. Load Dataset

```python
# TODO: update filename to match the actual file in data/raw/
data_file = RAW_DIR / "<filename>.csv"

df = pd.read_csv(data_file)
print(f"Loaded: {data_file}")
print(f"Shape : {df.shape}")
df.head()
```

## 2. Schema Validation

```python
schema_lines = ["# Schema Log\n"]
schema_lines.append(f"**Row count:** {len(df)}\n")
schema_lines.append(f"**Column count:** {len(df.columns)}\n")
schema_lines.append("\n## Columns\n")
schema_lines.append("| Column | dtype |\n|--------|-------|")

for col, dtype in df.dtypes.items():
    schema_lines.append(f"| {col} | {dtype} |")

schema_md = "\n".join(schema_lines)
(OUTPUT_DIR / "schema_log.md").write_text(schema_md)
print(schema_md)
```

## 3. Missingness Report

```python
missing_counts = df.isnull().sum()
missing_pct    = (missing_counts / len(df) * 100).round(2)
missing_df     = pd.DataFrame({"missing_count": missing_counts,
                                "missing_pct":   missing_pct})
missing_df     = missing_df[missing_df.missing_count > 0].sort_values(
                    "missing_pct", ascending=False)

report_lines = ["# Missingness Report\n"]
if missing_df.empty:
    report_lines.append("No missing values found.")
else:
    report_lines.append(missing_df.to_markdown())
    report_lines.append("\n## Handling Strategy\n")
    report_lines.append("<!-- TODO: justify every imputation / drop decision below -->")
    report_lines.append("\n| Column | Strategy | Justification |")
    report_lines.append("|--------|----------|---------------|")
    for col in missing_df.index:
        report_lines.append(f"| {col} | TODO | TODO |")

(OUTPUT_DIR / "missingness_report.md").write_text("\n".join(report_lines))
print(missing_df)
```
